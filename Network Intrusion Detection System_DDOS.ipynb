####Extracting DoS attacks from KDD Cup 1999 dataset using Python and CSV module"

```python
# Define the list of DoS attacks to extract
dos_attacks = {'back', 'land', 'neptune', 'pod', 'smurf', 'teardrop', 'normal'}

# Open the input and output files
with open('KDDCup99.csv', 'r') as input_file, open('pre_processed_data.csv', 'a', newline='') as output_file:

    # Create CSV reader and writer objects
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)

    # Iterate over the input rows
    for row in reader:
        
        # Check if the attack type is in the DoS attacks set
        if row[41] in dos_attacks:
            
            # Write the row to the output file
            writer.writerow(row)
```

####**MODEL-BUILDING**

####Building Gaussian Naive Bayes Classifier and performance Evaluation

Evaluation of Gaussian Naive Bayes Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns


# Fit a Gaussian Naive Bayes classifier to the training data
clf = GaussianNB(var_smoothing=0.000000010)
clf.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = clf.predict(x_test)

#train accuracy
train_acc = clf.score(x_train, y_train)
print("Train accuracy = {:.2f}%".format(train_acc*100))

# Calculate and print the accuracy, confusion matrix, and classification report
acc = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(acc * 100))

# Calculate the confusion matrix and visualize it using heatmap
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

print("Classification Report:\n", classification_report(y_test, y_pred))

```

```python
from sklearn.metrics import classification_report
import pandas as pd

report = classification_report(y_test, y_pred, output_dict=True)
df = pd.DataFrame(report).transpose()
df = df.drop('support', axis=1)  # drop the support column
df.style.background_gradient(cmap='Blues', axis=0, subset=['precision', 'recall', 'f1-score'])
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn import metrics
import matplotlib.pyplot as plt

fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, color='darkorange', lw=2, label='AUC = %0.2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
models.append("Naive Bayes Classifier")
scores.append(acc*100)
```

####Building Neural Network

Creating a neural network model using TensorFlow and evaluating its accuracy

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import numpy as np

# seed for reproducing same results
seed = 9
np.random.seed(seed)

# split the data into training (70%) and testing (30%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=seed)

# create the model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(5, input_dim=7, kernel_initializer='uniform', activation='relu'),
  tf.keras.layers.Dense(1, kernel_initializer='uniform', activation='sigmoid')
])

# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the model
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, batch_size=20, verbose=1)

# evaluate the model
_, accuracy = model.evaluate(x_test, y_test)
print("Accuracy: %.2f%%" % (accuracy * 100))



```
